```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(2,3, figsize=(20,15))
rms = pd.read_csv('rms_lig.txt', header=None)
ax[0,0].plot(rms, color='black')
ax[0,0].set_xlabel('etat', fontsize=15)
ax[0,0].set_ylabel('RMSD_ligand', fontsize=15)
ax[0,0].set_title('RMSD_ligand', fontsize=20)

rms_rec = pd.read_csv("rms_rec.txt", header=None)
ax[0,1].plot(rms_rec, color='black')
ax[0,1].set_xlabel('etat', fontsize=15)
ax[0,1].set_ylabel('RMSD_receptor', fontsize=15)
ax[0,1].set_title('RMS_receptor', fontsize=20)

rmsf = pd.read_csv("rmsf.txt", header=None)
ax[0,2].plot(rmsf, color='black')
ax[0,2].set_xlabel('etat', fontsize=15)
ax[0,2].set_ylabel('RMSF', fontsize=15)
ax[0,2].set_title('RMSF_1ps', fontsize=20)

rms_1ns = pd.read_csv('rms_lig_1ns.txt', header=None)
ax[1,0].plot(rms_1ns, color='black')
ax[1,0].set_xlabel('etat', fontsize=15)
ax[1,0].set_ylabel('RMSD_ligand', fontsize=15)
ax[1,0].set_title('RMSD_ligand_1ns', fontsize=20)

rms_rec_1ns = pd.read_csv("rms_rec_1ns.txt", header=None)
ax[1,1].plot(rms_rec_1ns, color='black')
ax[1,1].set_xlabel('etat', fontsize=15)
ax[1,1].set_ylabel('RMSD_receptor', fontsize=15)
ax[1,1].set_title('RMSD_receptor_1ns', fontsize=20)

rmsf_1ns = pd.read_csv("rmsf_1ns.txt", header=None)
ax[1,2].plot(rmsf_1ns, color='black')
ax[1,2].set_xlabel('etat', fontsize=15)
ax[1,2].set_ylabel('RMSF', fontsize=15)
ax[1,2].set_title('RMSF_1ns_1ns', fontsize=20)
plt.savefig('rmsf_1ns.png', dpi=500)
plt.show()
```
