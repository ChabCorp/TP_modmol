```python
import matplotlib.pyplot as plt
import pandas as pd

time = [i*0.002 for i in range(2000)]

E_tot = pd.read_csv('Etot_therm.txt', header=None)
E_kin = pd.read_csv('Ek_therm.txt', header=None)
E_pot = pd.read_csv('Ep_therm.txt', header=None)

fig, ax = plt.subplots(1,3, figsize=(20,5))
ax[0].plot(E_tot, label='E_tot', color='black')
ax[0].plot(E_kin, label='E_kin', color='red')
ax[0].plot(E_pot, label='E_pot', color='blue')
ax[0].set_xlabel('time', fontsize=15)
ax[0].set_ylabel('Energy', fontsize=15)
ax[0].set_title('Thermalization', fontsize=20)
ax[0].legend()

E_tot = pd.read_csv('Etot_eq.txt', header=None)
E_kin = pd.read_csv('Ek_eq.txt', header=None)
E_pot = pd.read_csv('Ep_eq.txt', header=None) 
ax[1].plot(E_tot, label='E_tot', color='black')
ax[1].plot(E_kin, label='E_kin', color='red')
ax[1].plot(E_pot, label='E_pot', color='blue')
ax[1].set_xlabel('time', fontsize=15)
ax[1].set_ylabel('Energy', fontsize=15)
ax[1].set_title('Equilibration', fontsize=20)
ax[1].legend()

E_tot = pd.read_csv('Etot_prod.txt', header=None)
E_kin = pd.read_csv('Ek_prod.txt', header=None)
E_pot = pd.read_csv('Ep_prod.txt', header=None) 
ax[2].plot(E_tot, label='E_tot', color='black')
ax[2].plot(E_kin, label='E_kin', color='red')
ax[2].plot(E_pot, label='E_pot', color='blue')
ax[2].set_xlabel('time', fontsize=15)
ax[2].set_ylabel('Energy', fontsize=15)
ax[2].set_title('Production', fontsize=20)
ax[2].legend()
plt.savefig('Energies.png', dpi=500)
plt.show()
```
